```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import os
from os import walk

import numpy as np
```

```python
train_imgs = "/content/gdrive/MyDrive/CV-DL/12_ObjectCategories_with_augmentations_1"

datagen = ImageDataGenerator(
        rotation_range=10, # rotation
        width_shift_range=0.2, # horizontal shift
        height_shift_range=0.2, # vertical shift
        zoom_range=0.2, # zoom
        horizontal_flip=True, # horizontal flip
        brightness_range=[0.2,1.2],
        shear_range=0.2,
        fill_mode = 'nearest',
        channel_shift_range=20.0) 


```

```python
for root, directories, files in os.walk(train_imgs, topdown=False):
  for file in files:

    img = load_img(os.path.join(root, file))  
    x = img_to_array(img)  
    x = x.reshape((1, ) + x.shape)  
    i = 0

    # generate 1 new augmented images 
    for batch in datagen.flow(x, batch_size = 1, 
                      save_to_dir =root,  
                      save_prefix = "aug_", save_format ='jpg'):
        i += 1
        if i > 0: 
            break

```
